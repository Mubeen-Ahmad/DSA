- ### 1 Insert node at beginning, end, Mid and any position.

```c++
#include <stdio.h>
#include <stdlib.h>

struct Node{
    int data;
    struct Node *next;
};

struct Node *head = NULL;
int count = 0;

void traverse(struct Node *n){
    struct Node *tmp = n;
    while(tmp != NULL){
        printf("%d\n",tmp->data);
        tmp = tmp->next;
    }
}

void add_first(int data){
        count += 1;
        struct Node *new_node = (struct Node *)malloc(sizeof(struct Node));
        new_node->data = data;
        
        if(head == NULL){
            head = new_node;
            head->next = NULL;
        }
        else{
            new_node ->next = head;
            head = new_node;
        }   
}

void add_end(int data){
        count += 1;
        struct Node *new_node = (struct Node *)malloc(sizeof(struct Node));
        new_node->data = data;
    
        if(head == NULL){
            head = new_node;
        }
        else{
            struct Node *tmp = head;
            while(tmp->next != NULL){
                tmp = tmp->next;
            }
            
            tmp->next = new_node;
            new_node->next = NULL;
        }   
}

void add_mid(int data,int top=0){
    count += 1;
    struct Node *new_node = (struct Node *)malloc(sizeof(struct Node));
    new_node->data = data; 
    
    if (head==NULL){
        head = new_node;
        head->next = NULL;
    }
    
    else if(head->next == NULL){
        head->next = new_node;
        new_node->next = NULL;
    }
    
    else if (count>2){
        // if divide-able
        if (count%2==1){
            
            int pos = 1;
            struct Node *tmp = head;
            while(tmp != NULL){
                if(count/2 == pos){break;}
                tmp = tmp->next;
                pos += 1;
                
            }
            new_node->next = tmp->next;
            tmp->next = new_node;
        
        }
        else{
            
            if (top){
                    int t_pos = 1;
                    struct Node *tmp = head;
                    while(tmp != NULL){
                        if(count/2 > t_pos){break;}
                        tmp = tmp->next;
                        t_pos += 1;
                
            }
                new_node->next = tmp->next;
                tmp->next = new_node;
                
            }
            else{
                
                 int x_pos = 1;
                    struct Node *tmp = head;
                    while(tmp != NULL){
                        if(count/2 == x_pos){break;}
                        tmp = tmp->next;
                        x_pos += 1;
                
            }
                new_node->next = tmp->next;
                tmp->next = new_node;
                
            }
            
        }
        
    }
    

}

void add_pos(int pos,int data){
    
    count += 1;
    struct Node *new_node = (struct Node *)malloc(sizeof(struct Node));
    new_node->data = data;
    
    struct Node *tmp = head;
    int tmp_pos = 1;
        
    while(tmp->next != NULL){
        
        printf("%d |%d|%d|\n",tmp->data,tmp_pos,pos);
        
        if(tmp_pos < pos){break;}
        tmp = tmp->next;
        tmp_pos += 1;
    }
    printf("\n\n%d",tmp->data);  

}
    
int main(){
    add_first(10);
    add_end(20);
    add_end(30);
    add_end(40);
    
    add_pos(2,121);
    
    // add_mid(30);
    
    
    // traverse(head);
}
```
